$$ y_1, y_2 \ge 0$$

```python
import numpy as np
from cylp.cy import CyCbcModel, CyClpSimplex
from cylp.py.modeling.CyLPModel import CyLPModel, CyLPArray

model = CyLPModel() # using the clp.py.modeling.CyLPModel modeling tool to create the LP

# add variables
x = model.addVariable('x', 3, isInt=True)
y = model.addVariable('y', 2)

A = np.matrix([[1., 2., 0],[1., 0, 1.]])
B = np.matrix([[1., 0, 0], [0, 0, 1.]])
D = np.matrix([[1., 2.],[0, 1]])
a = CyLPArray([5, 2.5])
b = CyLPArray([4.2, 3])
x_u= CyLPArray([2., 3.5])

model += A*x <= a
model += 2 <= B * x + D * y <= b
model += y >= 0
model += 1.1 <= x[1:3] <= x_u

c = CyLPArray([1., -2., 3.])

model.objective = c * x + 2 * y.sum()

s = CyClpSimplex(model) # Create the cylp.cy.CySimplex object, using the created model as input

cbcModel = s.getCbcModel() # Create the Cbc model, which is capable of solving MIPs using B&B

cbcModel.solve()
print(cbcModel.status)
sol_x = cbcModel.primalVariableSolution['x']
print(sol_x)
(abs(sol_x - np.array([0, 2, 2])) <= 10**-6).all()
```

We now create a data frame for all two-ingredient potions:

```python
def get_effects(ingredients):
    effects_list = []
    for ingredient in ingredients:
        effects = list(ingredients_df.loc[ingredients_df.name == ingredient, 
                                  ['effect_1', 'effect_2', 'effect_3', 'effect_4']].values[0])
        effects_list.append(effects)
        
    possible_effects = {}
    for effects in effects_list:
        for effect in effects:
            possible_effects[effect] = possible_effects.get(effect, 0) + 1
    
    created_effects = []
    for effect in possible_effects.keys():
        if possible_effects[effect] >= 2:
            created_effects.append(effect)
    
    return created_effects
```

```python
tic = time.perf_counter()
potion_df = pd.DataFrame()
potion_data = []
for i in range(len(ingredients_df)):
    toc = time.perf_counter()
    first = ingredients_df.name[i]
    print(f'Current First Ingredient: {first} \t Total elapsed time: {toc - tic:.2f} seconds')
    for j in range(i+1, len(ingredients_df)):
        second = ingredients_df.name[j]
        effects = get_effects([first, second])
        potion_data.append([first, second, 'NA', effects])
        for k in range(j+1, len(ingredients_df)):
            third = ingredients_df.name[k]
            effects = get_effects([first, second, third])
            potion_data.append([first, second, third, effects])
toc = time.perf_counter()
print(f'Total elapsed time: {toc - tic:.2f} seconds')
```

```python
potion_df = potion_df.append(potion_data, ignore_index = True)
```

The model is the following:

#### Parameters

$P = \{1, \dots, 121826\}$: Set of potions

$I = \{1, \dots, 109\}$: Set of ingredients

$I_p \subset{I}$: Set of ingredients required to make one unit of potion $p$ (note that $|I_p| \leq 3 \,\,\forall p \in P$)

$z_i \in \mathbb{Z}^+$: Amount of ingredient $i$ available

$c_p \in \mathbb{R}^+$: gold cost of potion $p$

#### Decision Variables

$y_p \in \mathbb{Z}^+$: Number of potions $p$ created

$x_{ip} \in \mathbb{Z}^+$: Amount of ingredient $i$ used to create potion $p$

#### Model

$$ \max \sum_{i \in p} c_p y_p$$
such that:
$$y_p \leq x_{ip} \qquad \forall i \in I_p,  \forall p \in P$$
$$\sum_{\{p \mid i \in I_p\}} x_{ip} \leq z_i \qquad \forall i \in I$$


An additional sophistication for the model is to allow the purchase of common ingredients, which would amount to adding a integer decision variable for each such ingredient, adding them to the RHSs of the second constraints, and subtracting the costs in the objective.
